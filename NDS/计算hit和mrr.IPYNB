```python
import pandas as pd
import numpy as np
from sklearn.metrics import matthews_corrcoef, balanced_accuracy_score

# ── 读取数据 ──────────────────────────────────────────────────────────────────
df = pd.read_csv("best_test_predictions.csv")
# 期望列: drug_index, adr_index, label, pred_score

# ── 全局指标：MCC & Balanced Accuracy（需要二值化预测）────────────────────────
# 以 0.5 为默认阈值，可按需调整
THRESHOLD = 0.5
y_true = df["label"].values
y_pred = (df["pred_score"] >= THRESHOLD).astype(int)

mcc      = matthews_corrcoef(y_true, y_pred)
bal_acc  = balanced_accuracy_score(y_true, y_pred)

# ── 按 drug 分组计算排序指标 ──────────────────────────────────────────────────
K_LIST = [1, 3, 5, 7, 10]

hit_lists = {k: [] for k in K_LIST}
mrr_list  = []

for drug_id, group in df.groupby("drug_index"):
    # 按预测分数降序排列，rank 从 1 开始
    group = group.sort_values("pred_score", ascending=False).reset_index(drop=True)
    group["rank"] = group.index + 1

    # 只看正样本
    pos = group[group["label"] == 1]
    if pos.empty:
        continue  # 该 drug 没有正样本，跳过

    best_rank = pos["rank"].min()  # 排名最靠前的正样本

    # Hit@K
    for k in K_LIST:
        hit_lists[k].append(1 if best_rank <= k else 0)

    # MRR: 1 / 最优正样本排名
    mrr_list.append(1.0 / best_rank)

# ── 汇总输出 ──────────────────────────────────────────────────────────────────
n_drugs = len(mrr_list)
print(f"药物数量（有正样本）: {n_drugs}")
print(f"二值化阈值         : {THRESHOLD}")
print("=" * 30)
for k in K_LIST:
    hit_k = np.mean(hit_lists[k])
    print(f"Hit@{k:<3}: {hit_k:.4f}")
print("-" * 30)
mrr = np.mean(mrr_list)
print(f"MRR              : {mrr:.4f}")
print("-" * 30)
print(f"MCC              : {mcc:.4f}")
print(f"Balanced Accuracy: {bal_acc:.4f}")
```
